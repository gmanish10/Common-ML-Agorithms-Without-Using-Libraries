```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import sklearn
from sklearn import datasets

iris = sklearn.datasets.load_iris()
x = iris.data[:, :2]
y = (iris.target != 0) * 1

x1 = np.full((150, 3), 1.0)
x1[:,1:] = x
x = x1

y = y.reshape(-1, 1)
```

```python
plt.scatter(x[:,1], x[:,2], c = y.ravel())
#plt.plot(x, sigmoid(-x.dot(w)))
```

```python
def sigmoid(z):
    return 1/(1+ np.exp(-z))
```

```python
def logistic_loss(y, y_hat):
    return -1*np.mean((y*np.log(y_hat) + (1-y)*(np.log(1-y_hat))))
```

```python
w = np.zeros((3, 1))
alpha = 0.01
```

```python
m = len(y)

for epoch in range(5000):
    z = np.matmul(x, w)
    A = sigmoid(z)
    loss = logistic_loss(y, A)
    dz = A - y
    dw = (1/m)*np.matmul(x.T, dz)
    
    w = w - alpha*dw
    
    if epoch%100 == 0:
        print(loss)
```
